```python
import pandas as pd


df = pd.read_csv('data/vgsales.csv')

print("JEUX SUR NES")
for game in df.values:
    if game[2] == "NES":
        print(game[1])

print("-----------------------")


print("TOTAL VENTE JAPON ")
somme_vente = 0
for game in df.values:
    somme_vente += float(game[8])
print(somme_vente)

print("-----------------------")

print("NOMBRE DE JEUX SORTIE EN 2008")
total = 0
for game in df.values:
    if float(game[3]) >= 2008:
        total += 1
print(total)

print("-----------------------")

print("POURCENTAGE JEUX WII")
total_jeux = 0
total_jeux_wii = 0
for game in df.values:
    if game[2] == "Wii":
        total_jeux_wii += 1
    total_jeux += 1
print(total_jeux_wii*100/total_jeux)

```
